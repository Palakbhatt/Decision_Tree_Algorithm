**IMPORTING LIBRARIES IN PYTHON**

```python
import numpy as np
import pandas as pd
import sklearn.datasets as datasets
```

**LOADING THE IRIS DATASET**

```python
iris=datasets.load_iris()
```

```python
#forming the iris dataframe
x=pd.DataFrame(iris.data, columns=iris.feature_names)
print(x.head(5))
```

```python
y=iris.target
y
```

**DESCRIBING THE DATASET**

**TRAINING AND SPLITTING THE DATASET**

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.33, random_state=42)
```

**DEFINING THE DECISION TREE ALGORITHM**

```python
from sklearn.tree import DecisionTreeClassifier
dtc=DecisionTreeClassifier()
dtc.fit(x, y)
print('Decision Tree Classifier Created')
```

```python
y_predict=dtc.predict(x_test)
y_predict
```

ACCURACY OF TRAINING DATA

```python
from sklearn.metrics import accuracy_score
dtc.fit(x_train, y_train)
predict=dtc.predict(x_train)
accuracy_score(predict, y_train)
```

**CONSTRUCTING CONFUSION MATRIX**

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test, y_predict)
```

**VISUALIZING DECISION TREE**

```python
import matplotlib.pyplot as plt
from sklearn import tree
```

```python
a=['sepal length(cm)', 'sepal width(cm)', 'petal length(cm)', 'petal width(cm)']
b=['setosa', 'versicolor', 'virginica']
fig, axis=plt.subplots(nrows = 1, ncols = 1, figsize = (4,4), dpi = 300)

tree.plot_tree(dtc, feature_names = a, class_names = b, filled = True);
```
